```python
import pyarrow.parquet as pq
df = pq.read_table('timeseries-2009-2016-no-datetime.parquet').to_pandas()
```

```python
df.head(5)
```

```python
dates = pq.read_table('timeseries-2009-2016-datetime.parquet').to_pandas()
dates.head(5)
```

```python
import pandas as pd
dates = pd.to_datetime(dates["Date Time"])
```

```python
set = pd.merge(dates, df, left_index=True, right_index=True)
set.head(5)
```

```python
set = set[set["Date Time"].dt.minute == 0]
```

```python
set.head(5)
```

```python
set_2012_2013 = set[set["Date Time"].dt.year.isin([2009, 2010, 2011, 2012])]
```
